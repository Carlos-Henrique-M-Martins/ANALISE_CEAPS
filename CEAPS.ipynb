# Criando uma base de dados unica

```python
despesas = pd.concat([dp_ceaps2019, dp_ceaps2020, dp_ceaps2021, dp_ceaps2022])
```

```python
despesas.head(-5)
```

```python
# Agora tem apenas uma tabela com todas as informações das anteriores. Isto facilitará e muito as analises 

print(f'Despesas tem {despesas.shape[0]} linhas e {despesas.shape[1]} colunas')
```

# Refinando a Base de dados

```python
# Informações do novo DataFrame

despesas.info()
```

```python
# Verificando se existe valores nulos

despesas.isnull().sum()
```

```python
#Transformando valores nulos em 0

despesas['DOCUMENTO'].fillna(value= 0, inplace = True)

# Pelo fato da coluna de  detalhamento poder conter informações relevantes para as analises foi optado em criar o termo de "não informado"

despesas['DETALHAMENTO'].fillna(value = 'Não informado', inplace = True)
```

```python
# Verificando se não existe mais valores nulos 

despesas.isnull().sum()
```

```python
# Eliminando colunas com informações não relacionadas a pesquisa

despesas = despesas.drop(columns = ['CNPJ_CPF', 'DOCUMENTO', 'COD_DOCUMENTO'])
```

```python
despesas
```

```python
# É preciso transformar a coluna data de object em uma datetime. Contudo existe algumas linhas dando erro então será necessário uma limpesa

despesas['NOVA_DATA'] = pd.to_datetime(despesas['DATA'], format="%d/%m/%Y", errors='coerce')
```

```python
despesas.query("NOVA_DATA == 'NaN'")
```

```python
despesas['DATA'] = despesas['DATA'].replace('31/01/0219','31/01/2019')
despesas['DATA'] = despesas['DATA'].replace('05/05/0202','05/05/2020')
despesas['DATA'] = despesas['DATA'].replace('29/10/0202','29/10/2020')
```

```python
despesas['DATA'] = pd.to_datetime(despesas['DATA'], format="%d/%m/%Y")
```

```python
despesas = despesas.drop(columns = 'NOVA_DATA')
```

```python
# A transformação da coluna data foi bem sucedida

despesas.info()
```

```python
# transformando ',' da coluna VALOR_REEMBOLSADO em '.' para tornala em float

despesas['VALOR_REEMBOLSADO'] = despesas['VALOR_REEMBOLSADO'].replace(',', '.', regex= True)

# Tranformando a coluna em float

despesas['VALOR_REEMBOLSADO'] = despesas['VALOR_REEMBOLSADO'].astype(float)
```

```python
# Mudando a nomenclatura da coluna tipo_despesa

despesas['TIPO_DESPESA'].unique()
```

```python
tipos_de_gastos = {'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.' : 'Aluguel de imoveis',
       'Divulgação da atividade parlamentar' : 'Divulgação da atividade parlamentar',
       'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ' : 'Aquisição de Materiais gerais',
       'Passagens aéreas, aquáticas e terrestres nacionais' : 'Transporte',
       'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar' : 'Contratações',
       'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes' : 'Verba de viagem',
       'Serviços de Segurança Privada' : 'Serviços de Segurança Privada'}
```

```python
despesas['TIPO_DESPESA'] = despesas['TIPO_DESPESA'].map(tipos_de_gastos)
```

```python
# Base de dados filtrada
despesas
```

CONCLUSÃO

Todas as alterações na tipologia dos dados, formatação do conteudo e exclusão de fatores não determinantes da pesquisa foram bem sucedidas.

# Analise da base de dados

## Qual é o periodo do ano em que mais ocorrre gastos publicos?


```python
# Visualisando o total de gastos por mes

ax = despesas.groupby('MES')['VALOR_REEMBOLSADO'].sum().reset_index()
fig = px.bar(ax, x='MES',y='VALOR_REEMBOLSADO', 
             title="Gastos por mes",
             text_auto='.2s'
            )
fig.update_traces(textfont_size=12, textangle=0, textposition="outside")
fig.show()
```

```python
# Visualizando os gastos em um ano eleitoral

ax1 = despesas.query("ANO == 2019")
ax = ax1.groupby('MES')['VALOR_REEMBOLSADO'].sum().reset_index()
fig = px.bar(ax, x='MES',y='VALOR_REEMBOLSADO', 
             title="Gastos por mes No ano de 2019", text_auto= '.2s'
            )
fig.update_traces(textfont_size=12, textangle=0, textposition="outside")
fig.show()
```

CONCLUSÃO

O primeiro grafico afirma que os gastos dos senadores crescem consideravelmente entre o fim e o inicio do ano (Desembro a março), sendo este o periodo que há mais gastos por parte dos Senadores. Contudo o segundo grafico mostra que em um ano eleitoral os gastos se concentrão em no meio para o final do ano muito provavelmente devido aos conpromissos de campanha.

## Qual a media de reenbousos feitos pelos senadores?


```python
# o numero do reembolso

despesas['VALOR_REEMBOLSADO'].sum()
```

```python
# E necessário criar uma media anual com o valor de reembolso com o numero de cadeiras do Senado (81) nos 4 anos analisados. 

media = (despesas['VALOR_REEMBOLSADO'].sum() / 81) / 4
media
```

```python
print(f'A media geral de reembouso anual de cada senador nesses ultimos 4 anos foi de R$ {media.round(2)}')
```

CONCLUSÃO 

Com a base de dados do CEAPS foi possivel definir que o reembolso medio de cada senador pede é de R$ 239.649,20 por ano.

## Qual e o gasto mais comum entre os senadores?

```python
gastos = despesas.groupby(['TIPO_DESPESA']).VALOR_REEMBOLSADO.sum()
gastos
```

```python
tipos_gastos = gastos.sort_values(ascending=False)
tipos_gastos
```

```python
tipos_gastos = pd.DataFrame(tipos_gastos)   
tipos_gastos
```

```python
tipos_gastos['VALOR_REEMBOLSADO'].sum()
```

CONCLUSÃO

Os maiores gastos por parte dos senadores são na contratação de pessoal, uso de transporte e na verba de viagem.
